```python
import pandas as pd
```

```python
import os
work_dir = os.getcwd()
df = pd.read_csv(work_dir + '/Documents/sDev/University/Datasets/penguins.csv')
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df['sex'] = df['sex'].replace('MALE', 1)
df['sex'] = df['sex'].replace('FEMALE', 0)
df
```

```python
df['island'].unique()
```

```python
dumies = pd.get_dummies(df['island'])
dumies
```

```python
df = df.drop('island', axis = 1)
```

```python
df['island_Biscoe'] = dumies['Biscoe']
df['island_Torgersen'] = dumies['Torgersen']
df['island_Dream'] = dumies['Dream']
```

```python
df
```

```python
df['species'].unique()
```

```python
df['species'] = df['species'].replace('Adelie', 0)
df['species'] = df['species'].replace('Chinstrap', 1)
df['species'] = df['species'].replace('Gentoo', 2)
#df = df.drop('species', axis = 1)
df
```

```python
X = df.drop('species', axis = 1)
```

```python
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
cls = DecisionTreeClassifier()
```

```python
cls.fit(train_x, train_y)
```

```python
y_pred = cls.predict(test_x)
```

```python
test_y.to_numpy()
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_pred, test_y))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "max_depth": [2, 3, 4, 5, 6, 7, 8, 9, 10],
    "min_samples_split": [2, 3, 4, 5, 6, 7, 8, 9, 10],
    "criterion": ['gini', 'entropy']
}
```

```python
grid_search = GridSearchCV(
    estimator = DecisionTreeClassifier(),
    param_grid = param_grid,
    cv = 10,
    scoring = 'accuracy'
)
```

```python
grid_search.fit(train_x, train_y)
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
best_pred_y = grid_search.best_estimator_.predict(test_x)
```

```python
print(classification_report(y_pred, test_y))
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_search = RandomizedSearchCV(
    estimator = DecisionTreeClassifier(),
    param_distributions = param_grid,
    cv = 10,
    scoring = 'accuracy'
)
```

```python
random_search.fit(train_x, train_y)
```

```python
random_search.best_score_
```

```python
random_search.best_params_
```

```python
from sklearn.model_selection import cross_validate
```

```python
cv_results = cross_validate(random_search.best_estimator_, X, y, cv = 10)
```
